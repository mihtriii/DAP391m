```python
import torch
import torch.nn.functional as F
import numpy as np
from sklearn.metrics import accuracy_score

# hyperparameters
examples = 1000
features = 100
hidden = 10
training_steps = 1000
lr = 0.1

X = np.random.randn(examples, features).astype(np.float32)
y = np.random.randint(0, 2, size=examples).astype(np.float32)

X = torch.tensor(X)
y = torch.tensor(y)

w1 = torch.randn(features, hidden, requires_grad=True)
b1 = torch.zeros(hidden, requires_grad=True)

w2 = torch.randn(hidden, requires_grad=True)
b2 = torch.tensor(0.0, requires_grad=True)

# prediction function
def forward(x):
    p1 = torch.tanh(x @ w1 + b1)
    p2 = torch.sigmoid(p1 @ w2 + b2)
    return p2

# accuracy trước training
with torch.no_grad():
    pred = (forward(X) > 0.5).float()
    print("Accuracy before Training:", accuracy_score(y.numpy(), pred.numpy()))

# training loop
for i in range(training_steps):

    p2 = forward(X)

    # binary cross entropy
    error = F.binary_cross_entropy(p2, y)

    # L2 regularization
    loss = error + 0.01 * (torch.sum(w1**2) + torch.sum(w2**2))

    loss.backward()

    # update parameters (gradient descent)
    with torch.no_grad():
        w1 -= lr * w1.grad
        b1 -= lr * b1.grad
        w2 -= lr * w2.grad
        b2 -= lr * b2.grad

        w1.grad.zero_()
        b1.grad.zero_()
        w2.grad.zero_()
        b2.grad.zero_()

# accuracy sau training
with torch.no_grad():
    pred = (forward(X) > 0.5).float()
    print("Accuracy after Training:", accuracy_score(y.numpy(), pred.numpy()))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Dataset
dataset = torch.tensor([
    [2.7810836,2.550537003,0],
    [1.465489372,2.362125076,0],
    [3.396561688,4.400293529,0],
    [1.38807019,1.850220317,0],
    [3.06407232,3.005305973,0],
    [7.627531214,2.759262235,1],
    [5.332441248,2.088626775,1],
    [6.922596716,1.77106367,1],
    [8.675418651,-0.242068655,1],
    [7.673756466,3.508563011,1]
], dtype=torch.float32)

X = dataset[:, :2]
y = dataset[:, 2].long()

# Model (2 inputs → 2 hidden → 2 outputs)
model = nn.Sequential(
    nn.Linear(2, 2),
    nn.Sigmoid(),
    nn.Linear(2, 2),
    nn.Sigmoid()
)

# Loss + optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.5)

# Training
epochs = 20
for epoch in range(epochs):

    outputs = model(X)
    loss = criterion(outputs, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f">epoch={epoch}, error={loss.item():.3f}")

for name, param in model.named_parameters():
    print(name, param.data)
```
