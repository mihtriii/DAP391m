```python
import Chefboost as chef
import pandas as pd

df = pd.read_csv('./dataset/golf.txt')
df.head()
# df.info()   
```

```python
print ("ID3 algorithm")

config = {'algorithm': 'ID3'}

print ("Training the model...")
model = chef.fit(df[:8].copy(), config)

xrow = ['Sunny', 'Hot', 'High', 'Weak']
prediction = chef.predict(model, xrow)
print ("Predicted class for the first row is: " + str(prediction))

c = 0
print("Comparing predictions with actual results...")
for index, instance in df[8:].iterrows():
    prediction = chef.predict(model, instance)
    actual = instance['Decision']
    print (f"Predicted: {prediction}, Actual: {actual}")
    if prediction == actual:
        c += 1
print ("Accuracy is: " + str(c/len(df[8:])*100) + "%")
```

```python
print ("C4.5 algorithm")
config = {'algorithm': 'C4.5'}
model = chef.fit(df[:8].copy(), config)
xrow = ['Sunny', 'Cool', 'High', 'Weak']
prediction = chef.predict(model, xrow)
print ("Predicted class for the first row is: " + str(prediction))
c = 0
for index, instance in df[8:].iterrows():
    prediction = chef.predict(model, instance)
    actual = instance['Decision']
    print (f"Predicted: {prediction}, Actual: {actual}")
    if prediction == actual:
        c += 1
print ("Accuracy is: " + str(c/len(df[8:])*100) + "%")
```

```python
print ("CART algorithm")

config = {'algorithm': 'CART'}
model = chef.fit(df[:8].copy(), config)

xrow = ['Sunny', 'Cool', 'High', 'Weak']
prediction = chef.predict(model, xrow)
print ("Predicted class for the first row is: " + str(prediction))

print("Comparing predictions with actual results...")
c = 0
for index, instance in df[8:].iterrows():
    prediction = chef.predict(model, instance)
    actual = instance['Decision']
    print (f"Predicted: {prediction}, Actual: {actual}")
    if prediction == actual:
        c += 1
print ("Accuracy is: " + str(c/len(df[8:])*100) + "%")
```

## Example 3 scikit-learn

```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.metrics import  classification_report, confusion_matrix
X, y = load_iris(return_X_y=True)
clf = tree.DecisionTreeClassifier()
tSize = int(X.shape[0] * 0.9)
X_train = X[:tSize, :]
y_train = y[:tSize]
X_test = X[tSize:, :]
y_test = y[tSize:]

clf = clf.fit(X_train, y_train)

tree.plot_tree(clf)

print("\nSklearn Decision Tree Classifier\n")
y_pred = clf.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Load golf dataset
df = pd.read_csv('./dataset/golf.txt')

# Encode categorical features
df_encoded = df.copy()
label_encoders = {}
for column in df_encoded.columns:
    if df_encoded[column].dtype == 'object':
        le = LabelEncoder()
        df_encoded[column] = le.fit_transform(df_encoded[column])
        label_encoders[column] = le

clf = tree.DecisionTreeClassifier()
tSize = int(df_encoded.shape[0] * 0.8)
X_train = df_encoded.iloc[:tSize, :-1]
y_train = df_encoded.iloc[:tSize, -1]
X_test = df_encoded.iloc[tSize:, :-1]
y_test = df_encoded.iloc[tSize:, -1]

clf = clf.fit(X_train, y_train)

tree.plot_tree(clf, feature_names=df.columns[:-1], class_names=[str(c) for c in df['Decision'].unique()], filled=True)

print("\nSklearn Decision Tree Classifier on Golf Dataset\n")
y_pred = clf.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"\nAccuracy: {clf.score(X_test, y_test)*100:.2f}%")
```

### Nhận xét Example 3 (Scikit-learn):
- Sklearn CART tối ưu hơn, cho metrics đầy đủ hơn
- Cần encode categorical features trước khi train
- Visualization rõ ràng, dễ phân tích decision path
