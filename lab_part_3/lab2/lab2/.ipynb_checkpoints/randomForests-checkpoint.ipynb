# Ensemble learning 

### Ensemble learning 
- machine learning paradigm where multiple models are used to solve the same problem
- voting scheme is then used to determine the class label for unlabeled instances
  - majority voting
  - veto voting (only single vote of classifier is required to mark email as spam)  
  - trust-based veto voting (includes trust of each classifier (vote's weight) )
  
### Most popular ensemble approaches
- boosting
- bagging
- stacking

### Boosting
- incremental process of building a sequence of classifiers
- each classifier works on the incorrectly classified instances of the previous one in the sequence.
- for example: AdaBoost 
  - introduced by Freund and Schapire (1997)
  - prone to overfitting
  
### Bagging (Bootstrap Aggregating) 
- building each classifier in the ensemble using a randomly drawn sample of the data
- each classifier has an equal vote when labeling unlabeled instances
- introduced by Leo Breiman in 1996 
- more robust than boosting against model overfitting

<b>Remark:</b> Sample is likely to have about 64% of instances appearing at least once in the sample.

### Stacking (stacked generalization)
- extends the cross-validation technique that partitions the data set into a held-in data set and a held-out data set
- trains the "base" models on the held-in data 
- uses "base" models outputs as inputs in the aggregator model

# Random Forests

### Introduction
- ensemble learning method 
- used for classification and regression
- developed by Leo Breiman in 2001
- combines Breiman's bagging sampling approach (1996), and Ho's random selection of features (1995) in order to construct a collection of decision trees with controlled variation

<img src="breiman.png">

### Applications
- ecology (species classification)
- medicine (response for breast cancer therapy)
- astronomy (astronomical object classification)
- traffic and transport planning (RF combined with K-d trees for traffic signs classification)
- agriculture (RF combined with SVM for crop classification)
- bioinformatics (developments survey in bioinformatics)

### How does Random Forests work?
- each tree in the ensemble acts as a base classifier to determine the class label of an unlabeled instance
- the class label with the most votes is used to classify the instance
- randomization is done also when selecting the best node to split on
- RF error rate depends on correlation and strength
    - increasing the correlation between any two trees in the RF increases the forest error rate
    - increasing the strength of the individual trees decreases the RF error rate

# Toy example

```python
#Predict the max temperature for tomorrow in the city using one year of past weather data
#data source: https://www.ncdc.noaa.gov/cdo-web/datasets

# Pandas is used for data manipulation
import pandas as pd
# Read in data and display first 5 rows
features = pd.read_csv('data.csv')
features.head(5)
```

```python
# Descriptive statistics for each column
features.describe()
```

```python
# One-hot encode the data using pandas get_dummies
features = pd.get_dummies(features)
# Display the first 5 rows of the last 12 columns
features.iloc[:,5:].head(5)
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(features['actual'])
print(labels)

# Remove the labels from the X features
# axis 1 refers to the columns
features= features.drop('actual', axis = 1)

# Saving feature names for later use
feature_list = list(features.columns)

# Convert to numpy array
features = np.array(features)
```

### Split into training/testing set

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

### Train the model

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```

### Prepare the forecasts 

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)


# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

### Calculate performance

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

### Visualize decision tree 

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot
# Pull out one tree from the forest
tree = rf.estimators_[5]


# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot


# Pull out one tree from the forest
tree = rf.estimators_[5]
# Export the image to a dot file
export_graphviz(tree, out_file = 'tree.dot', feature_names = feature_list, rounded = True, precision = 1)
# Use dot file to create a graph
(graph, ) = pydot.graph_from_dot_file('tree.dot')
# Write graph to a png file
graph.write_png('tree.png')
```

### Variable importance

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
# Import matplotlib for plotting and use magic command for Jupyter Notebooks
import matplotlib.pyplot as plt
%matplotlib inline
# Set the style
plt.style.use('fivethirtyeight')
# list of x locations for plotting
x_values = list(range(len(importances)))
# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')
# Tick labels for x axis
plt.xticks(x_values, feature_list, rotation='vertical')
# Axis labels and title
plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
```

### Extract two most important features

```python
# New random forest with only the two most important variables
rf_most_important = RandomForestRegressor(n_estimators= 1000, random_state=42)
# Extract the two most important features
important_indices = [feature_list.index('temp_1'), feature_list.index('average')]
train_important = train_features[:, important_indices]
test_important = test_features[:, important_indices]

# Train the random forest
rf_most_important.fit(train_important, train_labels)

# Make predictions and determine the error
predictions = rf_most_important.predict(test_important)
errors = abs(predictions - test_labels)

# Display the performance metrics
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
mape = np.mean(100 * (errors / test_labels))
accuracy = 100 - mape
print('Accuracy:', round(accuracy, 2), '%.')
```
